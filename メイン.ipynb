```python
"""
（´-`）.｡oO　　DOMを取得して加工するよー
"""


from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import chromedriver_binary

import pprint

import random

options = Options()
# options.add_argument('--headless')#headlessモードで実行するかどうか
driver = webdriver.Chrome("c:/Users/isao.sakuma/Downloads/chromedriver_win32/chromedriver.exe",options=options) 


# サイトにアクセス
driver.get('https://aiseki-ya.com/')



# htmlを取得・表示
print(driver.title)


# xpathを定義してfind関数で要素をリストで取得
xpath = '//*[@id="js-container"]/main/div[5]/div'
# xpath = '//*[@id="js-container"]/main/div[5]/div[1]/div[1]/p'
# xpath = '//*[@id="js-container"]/main/div[5]/div[2]/div[1]/p'
elems = driver.find_elements_by_xpath(xpath)

output = []  
count = 0
# 取得した要素を1つずつ表示
for elem in elems:
    data = elem.text.split()
    data[1] = data[1].split('/')
    data[2] = data[2].split('/')

    array = []
    array.append(data[0])
    # array.append(data[1][0].replace('人',''))
    array.append(random.randint(1, 30))
    array.append(data[1][1].replace('人',''))
    # array.append(data[2][0].replace('人',''))
    array.append(random.randint(1, 30))
    array.append(data[2][1].replace('人',''))

    output.append(array)

    count += 1


# ブラウザーを終了
driver.quit()

pprint.pprint(output)




```

```python
"""
（´-`）.｡oO　　表で出力するよー
"""

import numpy as np
import pandas as pd

df = pd.DataFrame(output, columns=['店名', '男性入店数', '男性席数', '女性入店数', '女性席数'])
df
```

```python
import matplotlib.pyplot as plt
import numpy as np
 
# フォント指定
plt.rcParams["font.family"] = "IPAexGothic"

height1 = df['女性入店数']
height2 = df['男性入店数']
 
left = np.arange(len(height1))  # numpyで横軸を設定
labels = df['店名']
 
plt.figure(figsize=(10,20))
height = 0.3
 
plt.barh(left, height1, color='m', height=height, align='center')
plt.barh(left+height, height2, color='b', height=height, align='center')
 

plt.yticks(left + height/2, labels)
plt.show()
```
